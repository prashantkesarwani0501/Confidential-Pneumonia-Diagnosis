```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score

# Pixelation function
def pixelate_image(img, block_size):
    h, w = img.shape
    temp = cv2.resize(img, (w // block_size, h // block_size), interpolation=cv2.INTER_LINEAR)
    pixelated = cv2.resize(temp, (w, h), interpolation=cv2.INTER_NEAREST)
    return pixelated

# Apply pixelation to dataset and save to new folder
def preprocess_and_save(input_dir, output_dir, block_size):
    os.makedirs(output_dir, exist_ok=True)
    for category in ['NORMAL', 'PNEUMONIA']:
        cat_path = os.path.join(input_dir, category)
        out_cat_path = os.path.join(output_dir, category)
        os.makedirs(out_cat_path, exist_ok=True)
        for img_name in os.listdir(cat_path):
            img_path = os.path.join(cat_path, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (128, 128))
            pixelated_img = pixelate_image(img, block_size)
            cv2.imwrite(os.path.join(out_cat_path, img_name), pixelated_img)

# Build CNN model
def build_model():
    model = models.Sequential([
        layers.Input(shape=(128, 128, 1)),
        layers.Conv2D(32, 3, activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(64, 3, activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Load datasets
def get_generators(path, batch_size=32):
    datagen = ImageDataGenerator(rescale=1./255)
    gen = datagen.flow_from_directory(
        path,
        target_size=(128, 128),
        color_mode='grayscale',
        class_mode='binary',
        batch_size=batch_size,
        shuffle=True
    )
    return gen

# Plot graphs
def plot_graphs(summary_df):
    sns.set(style="whitegrid")

    # Line plot: Accuracy vs Block Size
    plt.figure()
    sns.lineplot(data=summary_df, x='Block Size', y='Accuracy', marker='o')
    plt.title("Accuracy vs Pixelation Block Size")
    plt.savefig("privacy_utility_tradeoff_line.png")

    # Bar plot
    plt.figure()
    sns.barplot(data=summary_df, x='Block Size', y='Accuracy', palette='Blues_d')
    plt.title("Pixelation Accuracy Comparison")
    plt.savefig("pixelation_accuracy_bars.png")

    # Dual Axis Plot: Accuracy vs Privacy Score
    fig, ax1 = plt.subplots()
    sns.lineplot(data=summary_df, x='Block Size', y='Accuracy', marker='o', ax=ax1, color='b')
    ax2 = ax1.twinx()
    sns.lineplot(data=summary_df, x='Block Size', y='Privacy Score (%)', marker='o', ax=ax2, color='r')
    ax1.set_ylabel('Accuracy', color='b')
    ax2.set_ylabel('Privacy Score (%)', color='r')
    plt.title("Accuracy vs Privacy Score")
    plt.savefig("accuracy_vs_privacy.png")

    # Table plot
    fig, ax = plt.subplots()
    ax.axis('off')
    table = plt.table(cellText=summary_df.values, colLabels=summary_df.columns, loc='center', cellLoc='center')
    table.scale(1.2, 1.5)
    plt.title("Privacy-Utility Tradeoff Summary", fontsize=14)
    plt.savefig("privacy_utility_summary.png")

# Privacy Score Formula (simplified)
def compute_privacy_score(block_size):
    return min(100, block_size * 5)

# Main pipeline
base_path = "/Users/ayushsharma/Downloads/chest_xray"
results = []

for block_size in [5, 10, 15, 20, 25, 30]:
    print(f"Processing for block size: {block_size}")
    # Create pixelated data folders
    for split in ['train', 'val', 'test']:
        preprocess_and_save(os.path.join(base_path, split), f"pixelated_{block_size}/{split}", block_size)

    # Generators
    train_gen = get_generators(f"pixelated_{block_size}/train")
    val_gen = get_generators(f"pixelated_{block_size}/val")
    test_gen = get_generators(f"pixelated_{block_size}/test", batch_size=1)

    # Train model
    model = build_model()
    model.fit(train_gen, validation_data=val_gen, epochs=5, verbose=1)

    # Evaluate
    test_gen.reset()
    preds = model.predict(test_gen, verbose=0)
    pred_labels = (preds > 0.5).astype("int32")
    true_labels = test_gen.classes
    acc = accuracy_score(true_labels, pred_labels)

    # Privacy Score
    privacy_score = compute_privacy_score(block_size)
    info_loss = round((1 - acc / 0.85) * 100, 2)

    results.append({
        "Block Size": block_size,
        "Accuracy": round(acc, 4),
        "Privacy Score (%)": privacy_score,
        "Information Loss (%)": info_loss
    })

# Save results
summary_df = pd.DataFrame(results)
summary_df.to_csv("pixelation_summary.csv", index=False)
plot_graphs(summary_df)

print(summary_df)
```
